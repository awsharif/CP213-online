# Solving ordinary differential equations with Python

To solve the differential equation
\begin{align*}
\frac{dy}{dx} &= - x
\end{align*}
with the initial condition $y(0)=1$.

```python
from scipy.integrate import odeint
import numpy as np

def f(x, y, a):
    return -x

x_data = np.arange(0.0, 5.0, 0.01)
y0 = 1.0
a = 1.0

y_data = odeint(f, y0, x_data, args=(a,))


import pylab as plt

plt.plot(x_data, y_data)

plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')

plt.show()
```

```python
from scipy.integrate import odeint
import numpy as np

def f(x, y, a):
    res = []
    res.append(-a*x[0])
    res.append(-2.0*a*x[1])
    return res

x_data = np.arange(0.0, 5.0, 0.01)
y0 = [1.0, 2.0]
a = 1.0

solution = odeint(f, y0, x_data, args=(a,))

y_data = [[], []]
for vec in solution:
    y_data[0].append( vec[0] )
    y_data[1].append( vec[1] )

import pylab as plt

plt.plot(x_data, y_data[0])
plt.plot(x_data, y_data[1])

plt.show()
```
